```python
def find_outliers(df, column):
    """
    Calculates the outliers of a column in the dataframe.
    Outliers are values outside the accepted range, i.e values < Q1 - 1.5*IQR and values > Q3 + 1.5*IQR,
    where Q1 = 1st Quartile, Q2 = 2nd Quartile, Q3 = 3rd Quartile, IQR = InterQuartile Range = Q3-Q1
    """ 
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5*iqr
    upper_bound = q3 + 1.5*iqr
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

outliers_age = find_outliers(df, 'Age')
print("Outliers for Age:\n", outliers_age)

outliers_dpf = find_outliers(df, 'DiabetesPedigreeFunction')
print("Outliers for DiabetesPedigreeFunction:\n", outliers_dpf)
```

## 9. Is there a significant difference in the mean glucose levels between individuals with diabetes (Outcome = 1) and those without diabetes (Outcome = 0)?Formulate null and alternative hypotheses and employ a T-test to examine this relationship.Plot a histogram to visualize the results. Assume significance level as 0.05

```python
import scipy.stats as stats

# print("Difference in glucose levels between induviduals with daibetes and induviduals without diabetes =",mean_Withdiabetes-mean_Withoutdiabetes)
df.columns = df.columns.str.strip()

# Calculate mean glucose levels based on diabetes outcome
mean_Withoutdiabetes = df[df['Outcome'] == 0]['Glucose'].mean()
mean_Withdiabetes = df[df['Outcome'] == 1]['Glucose'].mean()

# Perform T-test
glucose_without_diabetes = df[df['Outcome'] == 0]['Glucose']
glucose_with_diabetes = df[df['Outcome'] == 1]['Glucose']
t_statistic, p_value = stats.ttest_ind(glucose_with_diabetes, glucose_without_diabetes)

# Print results
alpha = 0.05
print(f"Mean Glucose Without Diabetes: {mean_Withoutdiabetes}")
print(f"Mean Glucose With Diabetes: {mean_Withdiabetes}")
print(f"T-statistic: {t_statistic}, P-value: {p_value}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in mean glucose levels.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in mean glucose levels.")
```

```python
plt.figure(figsize=(10, 6))
plt.hist(glucose_without_diabetes, bins=30, alpha=0.5, label='Without Diabetes', color='blue')
plt.hist(glucose_with_diabetes, bins=30, alpha=0.5, label='With Diabetes', color='red')
plt.title('Histogram of Glucose Levels')
plt.xlabel('Glucose Level (mg/dL)')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```

 ## 10. Calculate the margin of error to quantify the precision of the analysis done previously and what you can infer from the results.

```python
std_Withoutdiabetes = glucose_without_diabetes.std()
std_Withdiabetes = glucose_with_diabetes.std()
n_Withoutdiabetes = len(glucose_without_diabetes)
n_Withdiabetes = len(glucose_with_diabetes)

z_score = 1.96 #for level of significance of 5% or 0.05
margin_of_error_Withoutdiabetes = z_score * (std_Withoutdiabetes / (n_Withoutdiabetes ** 0.5))
margin_of_error_Withdiabetes = z_score * (std_Withdiabetes / (n_Withdiabetes ** 0.5))

# Print Margin of Error results
print(f"Margin of Error Without Diabetes: {margin_of_error_Withoutdiabetes}")
print(f"Margin of Error With Diabetes: {margin_of_error_Withdiabetes}")
```

## 11. Perform a linear regression to predict 'DiabetesPedigreeFunction' using 'Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Age'. Validate the model using performance metrics like RMSE, MSE and R-squared value.

```python
x=df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','Age']]
y=df['DiabetesPedigreeFunction']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)#WE need to define the test and train split

from sklearn.linear_model import LinearRegression#IMport the linear regression module
Diabetes_Model=LinearRegression()
Diabetes_Model.fit(x_train,y_train)

#Define the prediction for Diabetes Pedigree Function using the features
from sklearn.metrics import mean_squared_error,r2_score

y_expected=Diabetes_Model.predict(x_test)

mse=mean_squared_error(y_test,y_expected)
print("THe mean squared error is: ",mse)

rmse=np.sqrt(mse)
print("The root mean squared error is: ",rmse)

r_squared=r2_score(y_test,y_expected)#basically observed and expected values

print("THe value for r^2 is calculated as: ",r_squared)
```
